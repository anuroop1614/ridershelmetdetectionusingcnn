```python
import os

os.makedirs("/content/dataset/helmet", exist_ok=True)
os.makedirs("/content/dataset/no_helmet", exist_ok=True)
```

```python
import os
import shutil
import xml.etree.ElementTree as ET

image_folder = "/content/images"
annotation_folder = "/content/annotations"

helmet_folder = "/content/dataset/helmet"
nohelmet_folder = "/content/dataset/no_helmet"

for file in os.listdir(annotation_folder):

    tree = ET.parse(os.path.join(annotation_folder, file))
    root = tree.getroot()

    image_name = root.find("filename").text
    src = os.path.join(image_folder, image_name)

    label = None

    # check labels in annotation
    for obj in root.findall("object"):
        name = obj.find("name").text.strip()

        if name == "Without Helmet":
            label = "no_helmet"
            break

        elif name == "With Helmet":
            label = "helmet"

    if label == "helmet":
        dst = os.path.join(helmet_folder, image_name)

    elif label == "no_helmet":
        dst = os.path.join(nohelmet_folder, image_name)

    else:
        continue

    if os.path.exists(src):
        shutil.copy(src, dst)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
```

```python
datagen = ImageDataGenerator(

    rescale=1./255,

    rotation_range=30,

    zoom_range=0.2,

    width_shift_range=0.2,

    height_shift_range=0.2,

    horizontal_flip=True,

    validation_split=0.2
)

train_data = datagen.flow_from_directory(

    "/content/dataset",

    target_size=(224,224),

    batch_size=32,

    class_mode="binary",

    subset="training"
)

val_data = datagen.flow_from_directory(

    "/content/dataset",

    target_size=(224,224),

    batch_size=32,

    class_mode="binary",

    subset="validation"
)
```

```python
base_model = MobileNetV2(

    input_shape=(224,224,3),

    include_top=False,

    weights='imagenet'
)

base_model.trainable = False
```

```python
model = models.Sequential([

    base_model,

    layers.GlobalAveragePooling2D(),

    layers.Dense(128, activation='relu'),

    layers.Dropout(0.5),

    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(

    optimizer='adam',

    loss='binary_crossentropy',

    metrics=['accuracy']
)
```

```python
history = model.fit(

    train_data,

    epochs=20,

    validation_data=val_data
)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

img_path = "/content/download_without.jpeg"

img = image.load_img(img_path, target_size=(224,224))

img_array = image.img_to_array(img)

img_array = img_array / 255.0

img_array = np.expand_dims(img_array, axis=0)

prediction = model.predict(img_array)

print(prediction)

if prediction < 0.5:
    print("Helmet Detected")
else:
    print("No Helmet")
```
